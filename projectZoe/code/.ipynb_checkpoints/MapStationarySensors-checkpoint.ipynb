```python
#Map grid in which to walk
constants = Constants()

walkingGrid = folium.Map(location=[55.943, -3.19], zoom_start=15, tiles="Stamen Toner")
#walkingArea = folium.Map(location=[55.943, -3.19], zoom_start=15)

gridResolution = constants.getGridSize()

#final corners
#corners = np.array([[55.94686, -3.19675], #NW
#                    [55.94686, -3.18133], #NE
#                    [55.93814, -3.18133], #SE
#                    [55.93814, -3.19675]]) #SW

corners = constants.getCorners()
staticCoords = constants.getStaticCoords()

topLat = corners[0][0]
bottomLat = corners[2][0]

leftLon = corners[0][1]
rightLon = corners[1][1]

height = topLat - bottomLat
width = rightLon - leftLon

print(height)
print(width)

heightInterval = height / gridResolution
widthInterval = width / gridResolution

#gridTuples = []

verticalLines = [] 

for i in range(gridResolution + 1):
    verticalLine = []
    topOfVerticalLine = tuple([topLat, leftLon + widthInterval * i])
    bottomOfVerticalLine = tuple([bottomLat , leftLon + widthInterval * i])
    verticalLine.append(topOfVerticalLine)
    verticalLine.append(bottomOfVerticalLine)
    folium.PolyLine(verticalLine, weight=1.5, opacity=0.8, color="#2E8B57").add_to(walkingGrid)
    
    horizontalLine = []
    leftOfHorizontalLine = tuple([bottomLat + heightInterval * i, leftLon])
    rightOfHorizontalLine = tuple([bottomLat + heightInterval * i, rightLon])
    horizontalLine.append(leftOfHorizontalLine)
    horizontalLine.append(rightOfHorizontalLine)
    folium.PolyLine(horizontalLine,  weight=4, opacity=1, color="#2E8B57").add_to(walkingGrid)
    
    #verticalLines.append(verticalLine)
    
#print(verticalLines)


#bottomOfVerticalLine = tuple([bottomLat + heightInterval * i, leftLong + widthInterval * j])

#cornerTuple = []
#for i in range(len(corners)):
#    cornerTuple.append(tuple([corners[i][0], corners[i][1]]))
    
#Add first point again to make square
#cornerTuple.append(tuple([corners[0][0], corners[0][1]]))
#folium.PolyLine(cornerTuple,  weight=1.5, opacity=0.8).add_to(walkingArea)


#plotter.plotPoints("walkingGrid", corners)#, labels)



#folium.PolyLine(cornerTuple).add_to(walkingGrid)

for i in range(staticCoords.shape[0]):
    #folium.CircleMarker((staticCoords[i][1], staticCoords[i][0]),
    folium.CircleMarker((staticCoords[i][0], staticCoords[i][1]),
                    radius=8,
                    #popup= staticLabels[i] + '<br>PM 2.5 = ' + str(staticPM2pt5[i]) + '<br>PM 10 = ' + str(staticPM10[i]),
                    color='#FF8C00',
                    weight=1.0,
                    #fill_color=assignColor(PM10_1306188AM[i]),
                    fill=True,
                    fill_opacity=1,
                   ).add_to(walkingGrid)
    
walkingGrid.save("../images/maps/finalWalkingGrid.html")
```
