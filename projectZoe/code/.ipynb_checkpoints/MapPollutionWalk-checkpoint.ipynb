```python
def createMap(mapObj, start_date, sid, PM_pred):
    end_date = getEndDate(start_date)
    data_dir = "/Users/zoepetard/Google Drive/Edinburgh/MscProj/FillingTheGaps/data/raw/personal/"+str(start_date)+"-"+str(end_date)+"/"
    sids = ['XXM007', 'XXM008']
    pdata = data_downloader.readAirSpeckPCSV(start_date, end_date, data_dir)
    belt_index = sids.index(sid)
     
    #for i in range(len(sids)):
    #maxPMwalk = np.max(pdata[belt_index]["PM2.5"])
    maxPM = np.max(PM_pred)
    #maxPM = np.max([maxPMwalk, maxPMpred])
    
    #minPMwalk = np.min(pdata[belt_index]["PM2.5"])
    minPM = np.min(PM_pred)
    #minPM = np.min([maxPMwalk, maxPMpred])
    
    ##Add prediction cells
    for num, cell in enumerate(cells):
        row = int(num / GRID_SIZE)
        col = int(num % GRID_SIZE)
        PM = PM_pred[row][col]
        cell_multi = cell[0]["geometry"]
        osm_reader.addCellPM(mapObj, cell_multi, PM, maxPM, minPM)
            
    ##Add validation walk
    for j in range(len(pdata[belt_index])):
        folium.CircleMarker((pdata[belt_index]["latitude"][j], pdata[belt_index]["longitude"][j]),
                    radius=5,
                    color='#000000',
                    weight=1.0,
                    fill_color=osm_reader.assignColor(pdata[belt_index]["PM2.5"][j], maxPM, minPM),
                    fill=True,
                    fill_opacity=1.0,
                   ).add_to(mapObj)

def getEndDate(start_date):
    if (start_date == 20180731):
        end_date = 20180801
    else:
        end_date = start_date + 1
    return end_date
```

```python
PM1 = []
PM2pt5 = []
PM10 = []

def setUp(start_date):
    #start_date = 20180719
    sid = 0
    end_date = getEndDate(start_date)
    data_dir = "/Users/zoepetard/Google Drive/Edinburgh/MscProj/FillingTheGaps/data/raw/personal/"+str(start_date)+"-"+str(end_date)+"/"
    sids = ['XXM007', 'XXM008']
    data_downloader = DataDownloader()

    pdata = data_downloader.readAirSpeckPCSV(start_date, end_date, data_dir)
    belt_index = 1 #sids.index(sid)
    
    PM1.append(findPMVar("PM1", pdata))
    PM2pt5.append(findPMVar("PM2.5", pdata))
    PM10.append(findPMVar("PM10", pdata))
        
    
        
def findPMVar(PMstr, pdata):
    belt_index = 1
    #print(len(pdata[belt_index]))
    if (len(pdata[belt_index]) == 0):
        belt_index = 0
    diff_sum = 0
    currentPM = pdata[belt_index][PMstr][0]
    for j in range(len(pdata[belt_index])):
        nextPM = pdata[belt_index][PMstr][j]
        diff = np.abs(currentPM - nextPM) / nextPM * 100
        diff_sum += diff
        currentPM = nextPM
        
    error = diff_sum / len(pdata[belt_index])
        
    print(PMstr +" " + str(error))
    return error 
    
    
constants = Constants()
dates = constants.getExperimentCollectedDates()

dates = [20180703, 20180704, 20180705, 20180709, 20180710, 20180716, 20180719, 20180723]

for date in dates:
    print(date)
    setUp(date)
    
#PM1clean = 
print("PM1 average: " + str(np.average(PM1)))
print("PM2.5 average: " + str(np.average(PM2pt5)))
print("PM10 average: " + str(np.average(PM10)))
#print(PM2pt5)
#print(PM10)

```
