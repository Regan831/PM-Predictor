```python
import os
from data_download import DataDownloader

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

from datetime import datetime
import time

from bokeh.io import show, output_notebook, output_file, reset_output
from bokeh.plotting import figure

import folium
```

```python
#Utility Functions
def toTimestamp(date):
    date = datetime.strptime(date.split(".")[0], "%Y-%m-%d %H:%M:%S")
    return time.mktime(date.timetuple())

def assignColorPM1(pm):
    if pm > 99999:
        return '#000000'
    elif pm < 1:
        color = '#1b7837'
    elif pm < 2:
        color ='#7fbf7b'
    elif pm < 3:
        color ='#d9f0d3'
    elif pm < 4:
        color ='#f7f7f7'
    elif pm < 5:
         color ='#e7d4e8'
    elif pm < 6:
        color = '#af8dc3'
    else:
        color ='#762a83'
        
    return color

def assignColorPM2pt5(pm):
    if pm > 99999:
        return '#000000'
    elif pm < 2:
        color = '#1b7837'
    elif pm < 4:
        color ='#7fbf7b'
    elif pm < 6:
        color ='#d9f0d3'
    elif pm < 8:
        color ='#f7f7f7'
    elif pm < 10:
         color ='#e7d4e8'
    elif pm < 12:
        color = '#af8dc3'
    else:
        color ='#762a83'
        
    return color

def assignColorPM10(pm):
    if pm > 99999:
        return '#000000'
    elif pm < 2:
        color = '#1b7837'
    elif pm < 5:
        color ='#7fbf7b'
    elif pm < 10:
        color ='#d9f0d3'
    elif pm < 15:
        color ='#f7f7f7'
    elif pm < 20:
         color ='#e7d4e8'
    elif pm < 25:
        color = '#af8dc3'
    else:
        color ='#762a83'
        
    return color
    
```

```python
start_date = 20180723
end_date = 20180724

uuids = ["02E5F77764B873DA",
        "200A7CED9D597407",
        "E5FD8C55EAA37555",
        "AA0E63CF5118F98F",
        "B61241EF668DBC2C",
        "E786F1568F65C296" ]

sids = ['XXM007', 'XXM008']

factors = [[1.0, 1.0, 1.0],
 [2.0100042,  1.54961648, 1.5126218, ],
 [1.59078671, 1.21618292, 1.19189916],
 [2.93987177, 2.3760729,  2.31180713],
 [2.75341775, 2.34367823, 2.21772871],
 [8.11136564, 7.08428589, 7.04657879],
 [5.05004303, 4.86456945, 4.72626118],
 [5.09431241, 4.90857996, 4.6303786 ]]

#calibration factors??

data_dir = "/Users/zoepetard/Google Drive/Edinburgh/MscProj/FillingTheGaps/data/raw/"+str(start_date)+"-"+str(end_date)+"/"

if not os.path.exists(data_dir):
    os.makedirs(data_dir)
```

```python
#Download data

#dataDownloader.init()
#data_dir = "/Users/zoepetard/Google Drive/Edinburgh/MscProj/FillingTheGaps/data/jun18/"
dataDownloader = DataDownloader()
dataDownloader.loadAirSpeckS(start_date, end_date, data_dir)
#retrieve @: data_dir + "/" + str(u) + '.csv'

dataDownloader.loadAirSpeckP(start_date, end_date, data_dir)
#retrieve @: data_dir + "/" + str(sid) + '_' + str(i) + '.csv'
#for i in range(start_date, end_date):
```

```python
#Read CSV

pdata = []
for i in range(len(sids)):
    pdata_by_day = []
    for j in range(start_date, end_date):
        pdata_by_day.append(pd.read_csv(data_dir + "/" + str(sids[i]) + '_' + str(j) + '.csv'))
        
    pdata_by_sid = pd.concat(pdata_by_day)
    pdata.append(pdata_by_sid)
        #pdata.append(pd.read_csv(data_dir + "/" + str(sids[i]) + '_' + str(j) + '.csv'))
        #sdata.append(sdatatemp)
    absoluteTime = []
    for k in range(len(pdata[i])):
        absoluteTime.append(toTimestamp(pdata[i]["Timestamp"].values[k]))
    pdata[i]["absoluteTime"] = absoluteTime
```

```python
#Plot Data

plot_names = ["PM1","PM2.5", "PM10"]#, "temperature", "humidity"]
CB_color_cycle = ['#377eb8', '#ff7f00', '#4daf4a', '#f781bf', '#a65628', '#984ea3', '#999999', '#e41a1c', '#dede00']

for pidx, plot_name in enumerate(plot_names):
    reset_output()
    output_file(data_dir + "/{}.html".format("ave_" + plot_name + "_" + str(start_date)))

    fig = figure(plot_width=1200, plot_height=600, title="Average " + plot_name + " (" + str(start_date) + ")")    
    #fig = figure(plot_width=900, plot_height=400, x_axis_type="datetime", title="Average " + plot_name + " (" + str(start_date) + ")")    

    for i in range(len(sids)):
        #fig.line(pdata[i].index, pdata[i][plot_name].values, legend='Personal: ' + sids[i], color=CB_color_cycle[len(uuids) + i])
        fig.line(pdata[i]["absoluteTime"].values, pdata[i][plot_name].values / factors[i + len(uuids)][pidx], legend='Personal: ' + sids[i], color=CB_color_cycle[len(uuids) + i])
        #fig.line(pdata[i]["absoluteTime"].values, pdata[i][plot_name].values, legend='Personal: ' + sids[i], color=CB_color_cycle[len(uuids) + i])
        print(factors[i + len(uuids)][pidx])
        
    fig.legend.location = "top_left"
    #fig.min_border_left = 30
    show(fig)
```

```python
from osm_reader import OSMReader

osm_reader = OSMReader()
osm_reader.init()
cells2 = osm_reader.getCellsWithMajorLU()
print(cells2)

#import folium
#pixelAndRoute = folium.Map(location=[55.943, -3.19], tiles="Stamen Toner", zoom_start=15)
PM_all = folium.Map(location=[55.94, -3.19], zoom_start=16)


for cell_num, cell in enumerate(cells2):
    cell_multi = cell[0]["geometry"]
    #cell_shape = shapely.geometry.asShape(cell_multi)
    #if "major_LU" in cell[0]["properties"]:
    major_LU_index = cell[0]["properties"]["major_LU"]
    osm_reader.addLandUse(PM_all, cell_multi, major_LU_index)
    
#PM_all.save("../images/maps/cycle9th/pixelAndRoute.html")
```

```python
#Plot Map

#PM_all = folium.Map(location=[55.94, -3.19], zoom_start=16)
      
        
#PM1.save("../images/maps/cycle9th/PM1.html")

#PM2pt5 = folium.Map(location=[55.94, -3.19], zoom_start=16)

#PM2pt5.save("../images/maps/cycle9th/PM2pt5.html")

#PM10 = folium.Map(location=[55.94, -3.19], zoom_start=16)

for i in range(len(sids)):
    for j in range(len(pdata[i])):
        folium.CircleMarker((pdata[i]["latitude"][j], pdata[i]["longitude"][j]),
                    radius=8,
                    #popup='Laurelhurst Park',
                    color='#000000',
                    weight=1.0,
                    fill_color=assignColorPM10(pdata[i]["PM10"][j]),
                    fill=True,
                    fill_opacity=1.0,
                   ).add_to(PM_all)
        
        
for i in range(len(sids)):
    for j in range(len(pdata[i])):
        folium.CircleMarker((pdata[i]["latitude"][j], pdata[i]["longitude"][j]),
                    radius=5,
                    #popup='Laurelhurst Park',
                    color='#000000',
                    weight=1.0,
                    fill_color=assignColorPM2pt5(pdata[i]["PM2.5"][j]),
                    fill=True,
                    fill_opacity=1.0,
                   ).add_to(PM_all)
        
        
for i in range(len(sids)):
    for j in range(len(pdata[i])):
        folium.CircleMarker((pdata[i]["latitude"][j], pdata[i]["longitude"][j]),
                    radius=2,
                    #popup='Laurelhurst Park',
                    color='#000000',
                    weight=1.0,
                    fill_color=assignColorPM1(pdata[i]["PM1"][j]),
                    fill=True,
                    fill_opacity=1.0,
                   ).add_to(PM_all)
        
        
PM_all.save("../images/maps/PM_all_Jul6.html")
```
